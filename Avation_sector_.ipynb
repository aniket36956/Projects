# Handle Outlires using IQR Technique

```python
numeric_col=['Age','Flight Distance','Departure Delay in Minutes','Arrival Delay in Minutes']
plt.figure(figsize=(12,12))
for i,columns in enumerate(numeric_col,1):
    plt.subplot(3,3,i)
    sns.boxplot(x=df[columns])
    plt.title(columns)
plt.tight_layout()
plt.show()
```

```python
q1=df['Flight Distance'].quantile(0.25)
q3=df['Flight Distance'].quantile(0.75)
print('q1',q1)
print('q3',q3)

iqr=q3-q1
print(iqr)

lb=q1-1.5*iqr
ub=q3+1.5*iqr
print('lb',lb)
print('ub',ub)
```

```python
df=df[(df['Flight Distance']>lb) & (df['Flight Distance']<4000)]
```

```python
plt.figure(figsize=(3,3))
sns.boxplot(x=df['Flight Distance'])
plt.show()
```

```python
q1=df['Departure Delay in Minutes'].quantile(0.25)
q3=df['Departure Delay in Minutes'].quantile(0.75)
print('q1',q1)
print('q3',q3)

iqr=q3-q1
print(iqr)

lb=q1-1.5*iqr
ub=q3+1.5*iqr
print('lb',lb)
print('ub',ub)
```

```python
df=df[(df['Departure Delay in Minutes']>lb) & (df['Departure Delay in Minutes']<ub)]
```

```python
plt.figure(figsize=(3,3))
sns.boxplot(x=df['Departure Delay in Minutes'])
plt.show()
```

```python
q1=df['Arrival Delay in Minutes'].quantile(0.25)
q3=df['Arrival Delay in Minutes'].quantile(0.75)
print('q1',q1)
print('q3',q3)

iqr=q3-q1
print(iqr)

lb=q1-1.5*iqr
ub=q3+1.5*iqr
print('lb',lb)
print('ub',ub)
```

```python
df=df[(df['Arrival Delay in Minutes']>lb) & (df['Arrival Delay in Minutes']<ub)]
```

```python
plt.figure(figsize=(3,3))
sns.boxplot(x=df['Arrival Delay in Minutes'])
plt.show()
```

# Handle Imblanced Data 

```python
# Synthetic Minority Over-sampling Technique (SMOTE), which is used for balancing imbalanced datasets
from imblearn.over_sampling import BorderlineSMOTE
```

```python
X=df.drop('satisfaction',axis=1)
y=df['satisfaction']
```

```python
smote=BorderlineSMOTE()
```

```python
X,y=smote.fit_resample(X,y)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=34)
```

```python
X_train.shape
```

```python
y_train.shape
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
# Fit the model
rf.fit(X_train,y_train)
```

```python
rf.predict(X_test)
```

```python
rf.predict([[0.0,1.0,67.0,1.0,0.0,680.0,3.0,3.0,3.0,3.0,2.0,4.0,5.0,3.0,3.0,3.0,3.0,3.0,3.0,4.0,2.0,0.0]])
```

```python
# Make predictions on the test set
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('accuracy_score',accuracy_score(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

# Receiver Operating Characteristic (ROC) Curve

```python
# The ROC curve and the Area Under the Curve (AUC) are commonly used metrics in binary classification problems.Eg.Imbalance dataset


from sklearn.metrics import roc_curve, auc

# Assuming y_test contains the true labels and y_pred contains the predicted probabilities or decision values
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

print(roc_auc)

# fpr: False Positive Rate at threshold values.
# tpr: True Positive Rate (Sensitivity or Recall) at  threshold values.
# thresholds: Threshold values used to calculate the corresponding FPR and TPR.
```

```python
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr,color='darkorange',label=f'ROC curve(AUC={roc_auc:.2f})')
plt.plot([0, 1],[0, 1], color='navy',linestyle='--',label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
