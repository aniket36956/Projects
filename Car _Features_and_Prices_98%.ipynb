# IQR Technique

```python
q1=df['Year'].quantile(0.25)
q3=df['Year'].quantile(0.75)
print('q1',q1)
print('q3',q3)

iqr=q3-q1
print(iqr)

lb=q1-1.5*iqr
ub=q3+1.5*iqr
print('lb',lb)
print('ub',ub)
```
